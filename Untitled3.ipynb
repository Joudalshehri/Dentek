```python
import yaml

data_config = {
    'path': '/content/Final_Split_Dataset',
    'train': 'train/images',
    'val': 'val/images',
    'test': 'test/images',
    'nc': 3,
    'names': ['Impacted_tooth', 'Missing_teeth', 'Bone_Loss']
}

with open('dental_final.yaml', 'w') as f:
    yaml.dump(data_config, f)

print("✅ ملف الـ YAML جاهز للتدريب!")
```

```python
import os

# المسارات في جوجل كولاب
train_labels_path = '/content/Final_Split_Dataset/train/labels'
val_labels_path = '/content/Final_Split_Dataset/val/labels'

def clean_labels(directory):
    if not os.path.exists(directory):
        print(f"المسار غير موجود: {directory}")
        return

    count = 0
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r') as f:
                lines = f.readlines()

            # الاحتفاظ فقط بالكلاسات 0، 1، 2
            filtered_lines = [line for line in lines if int(line.split()[0]) < 3]

            with open(file_path, 'w') as f:
                f.writelines(filtered_lines)
            count += 1
    print(f"تم تنظيف {count} ملف في {directory}")

clean_labels(train_labels_path)
clean_labels(val_labels_path)
```

```python
from ultralytics import YOLO

# تحميل أوزان YOLOv8m (بما أنكِ كنتِ تستخدمين النسخة المتوسطة m)
model = YOLO('yolov8m.pt')


results = model.train(
    data='dental_final.yaml',
    epochs=100,
    imgsz=1024,
    batch=8,
    project='/content/drive/MyDrive/Dental_Expert_Project', # الحفظ في الدرايف
    name='Dental_Expert_Final'
)
```
